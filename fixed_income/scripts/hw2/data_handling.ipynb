```python
import pandas as pd
import numpy as np

static_path = rf"C:\Users\anjal\Desktop\Fordham\Spring 2026\Interest Rate Derivatives\Homework\fixed_income\data\hw2_data\hw2_bond_info.xlsx"
pricing_path = rf"C:\Users\anjal\Desktop\Fordham\Spring 2026\Interest Rate Derivatives\Homework\fixed_income\data\hw2_data\hw2_bond_pricing_data.xlsx"
```

```python
bond_info = pd.read_excel(static_path)
pricing_info = pd.read_excel(pricing_path, header=[0])
```

```python
pricing_info.head()
```

```python
df = pricing_info.copy()   # the fixed wide table 

df = df.rename(columns={"dates": "date"})
df["date"] = pd.to_datetime(df["date"], dayfirst=True, errors="coerce")

# wide -> long
pricing_long = df.melt(
    id_vars="date",
    var_name="id",        # like "91282CPJ Govt"
    value_name="price"
)

# clean
pricing_long["id"] = pricing_long["id"].astype(str).str.strip()
pricing_long["price"] = pd.to_numeric(pricing_long["price"], errors="coerce")

print("Unique IDs in pricing_long:", pricing_long["id"].nunique())
pricing_long.head()
```

```python
pricing_long.to_csv(r"C:\Users\anjal\Desktop\Fordham\Spring 2026\Interest Rate Derivatives\Homework\fixed_income\data\hw2_data\hw2_bond_pricing_cleaned.csv", index=False)
```
